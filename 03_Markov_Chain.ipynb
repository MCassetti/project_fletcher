```python
import pickle
import numpy as np
import random
import re
import string
from nltk import sent_tokenize

cleaned_sentances = []
with open('cleaned_sentances_updated.pkl','rb') as fp:
    cleaned_sentances = pickle.load(fp)
```

```python
def limiter(x, limit):
    for i in range(len(x)):
        if i >= limit and x[i] == " ":
            break
    return x[:i]
# def clean_tweets(tweets):
#     cleaner_tweets = []
#     tweets = sent_tokenize(tweets)
#     if len(tweets) > 1: 
#         for tweet in tweets:
#             if (re.match('[?.!]', tweet[-1]) is not None):
#                 print(tweet[-1])
        
# #         for word in tweets:
# #             if (re.match('', tweets[-1]) is not None):
# #                 print(tweets[-1])
# #                 tweet_components.append(word)
# #             cleaner_tweets.append(' '.join(tweet_components))
#     else:
#         cleaner_tweets = tweets
        
#     return cleaner_tweets

def generate_text_bigram_seeded(following_dict, bigram_dict, num_words,seeds):
    # randomly select a word.
    # Ideas for improvements: 
    # - could mark words as sentence starters
    # - could weight by frequency.
    
    all_bigrams = list(bigram_dict.keys())
    current_bigram = random.choice(seeds)
    text = [current_bigram]
    for _ in range(num_words - 1):
        if current_bigram in all_bigrams:
            
            next_word = random.choice(bigram_dict[current_bigram])
        else:
            if following_dict[text[-1]]:
                next_word = random.choice(following_dict[text[-1]])
            else:
                continue
        text.append(next_word)
        current_bigram = (current_bigram[1], next_word)
    return ' '.join(text)

def create_dictionary(cleaned_sentances):
    following = {}
    count = {}
    for words in cleaned_sentances:
        words = words.split()
        words = [x for x in words if x]
        for index, word in enumerate(words[:-1]):
            # ensure the word is in the dictionary
            if word not in following:
                following[word] = []
                count[word] = 0

            # append the next word
            following[word].append(words[index + 1])
            count[word] += 1

        if words:
            last_word = words[-1]
            if last_word not in following:
                following[last_word] = []
                count[last_word] = 0
    return count, following

def create_bigrams(cleaned_sentances):
    bigrams = {}
    for words in cleaned_sentances:
        words = words.split()
        words = [x for x in words if x]
        for w1, w2, w3 in zip(words[:-2], words[1:-1], words[2:]):
            bigram = (w1,w2)
            if bigram not in bigrams:
                bigrams[bigram] = [w3]
            else:
                bigrams[bigram].append(w3)   
    return bigrams
```

```python
count, following = create_dictionary(cleaned_sentances)
bigrams = create_bigrams(cleaned_sentances)
```

```python
seeds = ['bitcoin','ethereum','litecoin','satoshi','charlie','ripple','stellar','zcash','doge','shiba','lambos','moon','ltc','btc','when','hodl','why']
markov_tweets = []
for _ in range(0,100):
    wordlist = generate_text_bigram_seeded(following, bigrams, 280,seeds)
    tweet = limiter(wordlist,180)
    markov_tweets.append(tweet)
    print(tweet)
    print('\n')
```
