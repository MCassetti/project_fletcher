```python
%time
import praw
import json

with open('reddit.json', 'r') as f:
    config = json.load(f)

r = praw.Reddit(client_id=config['client_id'],
                client_secret=config['client_secret'],
                user_agent=config['user_agent'])
with open('output_comments_query_crypto_all_plus.txt', 'w+') as handle:
#     subreddits = ['Bitcoin','CryptoCurrency','Ethereum','bitcoinmemes',\
#                   'Litecoin','AltCoin','Dogecoin','CryptoMarkets',\
#                   'btc','bitcoinxt','bitcoinall', 'BitcoinMarkets'\
#                    ,'bitcoin_uncensored','BTCuncensored','WeAreBitcoin',\
#                    'BetterBitcoin','CoinMarketCap','EthereumClassicx2',\
#                    'Reddcoin','Ripple','Bitcoincash','Stellar',\
#                     'eos','zec','monero','Tronix','Dash','tether']
    for sub in subreddits:
        print(sub)
        try:
            subreddit = r.subreddit(sub)  
            #top_posts = subreddit.hot(limit=None)
            all_comments = subreddit.comments(limit=None)
            #print(subreddit.comments(limit=None))
            for comment in all_comments:
                handle.write(comment.body)  
        except:
            print('something went wrong...')
```
